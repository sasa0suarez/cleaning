Importo las bibliotecas necesarias para la visualización de datos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Selección de columnas**: x = df2[['area','salary']] selecciona las columnas ‘area’ y ‘salary’ de tu DataFrame df2 y las guarda en un nuevo DataFrame x.

**Agrupación y suma**: x = x.groupby('area')['salary'].sum().reset_index() agrupa el DataFrame x por la columna ‘area’, suma los valores de ‘salary’ para cada ‘area’, y luego resetea el índice del DataFrame resultante.

**Ordenación**: x = x.sort_values(by='salary',ascending=False) ordena el DataFrame x en orden descendente por la columna ‘salary’.

```python
x = df2[['area','salary']]
x = x.groupby('area')['salary'].sum().reset_index()
x = x.sort_values(by='salary',ascending=False)
x
```

```python
plt.pie(x['salary'], radius=3, autopct='%.2f%%')
plt.legend(labels=x['area'], loc='upper left', bbox_to_anchor=(1, 0, 0.5, 1))
plt.axis("equal")
plt.show()

plt.barh(x['area'], x['salary'])
plt.xlabel('Salary cientos de millones')
plt.ylabel('Area')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plot = sns.barplot(x='area', y='salary', data=x, palette='rocket')
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
```
